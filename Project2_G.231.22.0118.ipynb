```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
data = pd.read_csv('curahhujan.csv')
```

```python
X = np.arange(1,13)
fig = plt.figure(figsize=(13, 5))

ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Bulan')
ax.set_ylabel('Curah Hujan (mm)')
ax.bar(X + 0.00, data['1981-1990'], color = 'blue', width = 0.25)
ax.bar(X + 0.25, data['1991-2000'], color = 'red', width = 0.25)
ax.bar(X + 0.50, data['2001-2010'], color = 'green', width = 0.25)
ax.legend(['1981-1990', '1991-2000', '2001-2010'], ncol=3, loc='upper center', frameon=False)

for d in [100, 200, 300, 400, 500, 600, 700]:
  if d != 700:
    plt.axhline(y=d, color='black', linestyle='-', linewidth=0.3)
  else:
    plt.axhline(y=d, color='black', linestyle=' ', linewidth=0.3)

ax.set_xticks(X + 0.25)
ax.set_xticklabels(np.arange(1, 13))

plt.show()
```

```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
data1 = [23, 25.2, 25.5, 24.8, 22.5, 22.2, 23, 22, 25.9]
data2 = [23.5, 25.3, 25.1, 25.5, 24.9, 25, 24.5, 24, 26]
data3 = [24, 25, 23.8, 23.9, 24, 24.1, 24, 22.9, 23]
dates = pd.date_range(start='2013-07-24', periods=9, freq='3H')

plt.figure(figsize=(13, 5))

plt.plot(dates, data1, marker='o', color='blue', label='Asimilasi')
plt.plot(dates, data3, marker='o', color='yellow', label='Observasi')
plt.plot(dates, data2, marker='o', color='red', label='Non Asimilasi')

plt.xlabel('Waktu')
plt.ylabel('Celcius')
plt.title('Suhu Permukaan')
plt.legend(loc="lower right")

for d in [20,21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M UTC'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())


plt.show()
```
